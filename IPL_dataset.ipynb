```python
# creating a new column based on the values
cr = ['CR(R)', 'Cr', 'Cr(R)', 'Crore', 'Crores', 'Crores(R)','crore(R)', 'crores(R)']
lakh = ['Lakh', 'Lakh(R)', 'Lakhs', 'Lakhs(R)','lakh', 'lakh(R)', 'lakhs(R)']
cr_value = 10000000
lakh_value = 100000
```

```python
for i,j in enumerate(ipl_team["Auction Price"]):
    x = j.split()
    y = x[-1]
    z = x[1]
    if y in cr:
        ipl_team["Auction Price"][i] = float(z)*cr_value
    elif y in lakh:
        ipl_team["Auction Price"][i] = float(z)*lakh_value
    else:
        ipl_team["Auction Price"][i] = 0
                
```

### match played by each team

```python
team1 =pd.DataFrame(match_info.groupby(by ="team1")["id"].agg("count"))
team2 =pd.DataFrame(match_info.groupby(by ="team2")["id"].agg("count"))
total_match_played =pd.concat([team1,team2],axis =0).reset_index().groupby(by = "index")["id"].agg("sum").reset_index().sort_values("index")
total_match_played
```

```python
winner = match_info.groupby(by ="winner").agg("count").sort_values(by = "id",ascending = False)["id"].reset_index()
winner = winner[winner.winner != "N/A"].sort_values(by ="winner").reset_index()
winner.drop("index",axis =1,inplace = True)
winner
```

```python
winner["win_percentage"] = winner["id"]/total_match_played["id"]*100
```

### top 5 teams who has played more than 50 matches.

```python
winner[winner.id > 50].sort_values("win_percentage",ascending = False).head()
```
